```python
# import libraries
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# import data 
df=pd.read_csv('data.csv',index_col=0,encoding='utf-8')
```

```python
df
```

```python
# data preprocessing 
df.drop('defaultAudioLanguage',inplace=True,axis=1,errors = 'ignore')
df.drop('defaultLanguage',inplace=True,axis=1,errors = 'ignore')
df.drop('liveBroadcastContent',inplace=True,axis=1,errors = 'ignore')
df.drop('localized',inplace=True,axis=1,errors = 'ignore')
```

```python
# data preprocessing 
df.isnull().sum()
```

```python
# data preprocessing 
for i in df.columns:
    if i=='tags' :
        df[i].fillna('non',inplace=True)
    else:
        df[i].fillna(0,inplace=True)
        
```

```python
df.isnull().sum()
```

```python
# data preprocessing 
df['publishedAt']=pd.to_datetime(df['publishedAt'],format='%Y-%m-%dT%H:%M:%S.%fZ')
```

```python
def YTDurationToSeconds(duration):
  match = re.match('PT(\d+H)?(\d+M)?(\d+S)?', duration).groups()
  hours = _js_parseInt(match[0]) if match[0] else 0
  minutes = _js_parseInt(match[1]) if match[1] else 0
  seconds = _js_parseInt(match[2]) if match[2] else 0
  return hours * 3600 + minutes * 60 + seconds
def _js_parseInt(string):
    return int(''.join([x for x in string if x.isdigit()]))
```

```python
import re
df['duration']=df['duration'].apply(YTDurationToSeconds)
```

```python
# apply  category names
df['categoryId']=df['categoryId'].astype(str)
df["Category"]=df["categoryId"].map(category_names)
```

```python
df['publish_year']= df['publishedAt'].dt.year
df['publish_month']= df['publishedAt'].dt.month
df['publish_day']= df['publishedAt'].dt.day
df['publish_hour']= df['publishedAt'].dt.hour
```

```python
metrics = ['likeCount','viewCount','commentCount','dislikeCount']


grp_cat = df.groupby('categoryId')[metrics].sum()
grp_cat.reset_index(inplace=True)
```

```python
grp_cat['category']=grp_cat['categoryId'].map(category_names)
```

```python
grp_cat.sort_values(metrics,ascending=False)
```

```python
grp_cat.plot(kind='barh',x='category',y='likeCount',title=' category likes')
```

```python
grp_cat.plot(kind='barh',x='category',y='viewCount',title=' category views')
```

```python
grp_cat.plot(kind='barh',x='category',y='commentCount',title=' category comments')
```

```python
grp_cat.plot(kind='barh',x='category',y='dislikeCount',title=' category dislikes')
```
